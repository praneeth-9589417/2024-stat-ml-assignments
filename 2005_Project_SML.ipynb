#Logistic Regression.

**Importing all the necessary modules for training the suitable data**

```python
import sklearn
import sklearn.linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report
import seaborn as sns
```

**Training the model**

```python
'''Training the model.'''
X_train, X_test, y_train, y_test = train_test_split(x_array, y_array, test_size=0.2, random_state=False)
loR = LogisticRegression()
```

**Fitting the model.**

```python
loR.fit(X_train, y_train)
```

**y predicted value**

```python
y_pred = loR.predict(X_test)
```

**Print the output**

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: \n{accuracy * 100:.2f}")
print()
print("Confusion Matrix: ")
print(confusion_matrix(y_test, y_pred))
print()
print("Precision: ")
print(precision_score(y_test, y_pred)*100)
print()
print("\nClassification Report: ")
print(classification_report(y_test, y_pred))
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report

sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

**Fitting and predicting the value.**

```python
lr = LogisticRegression()
model = lr.fit(X_train_res, y_train_res)
y_pred = model.predict(X_test)
```

#SVM - Support Vector Machine.

**Importing the Module**

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

**Making Classification**

```python
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)
```

**Creating the training and testing data**

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Standard Scaling the training and testing data**

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Creating the SVM model from sk-learn**

```python
svm = SVC(kernel='linear', random_state=42)
```

**Fitting the value**

```python
svm.fit(X_train, y_train)
```

**Predicting the y value**

```python
y_pred = svm.predict(X_test)
```

**Printing the outputa**

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: \n{accuracy * 100:.2f}")
print()
print("Confusion Matrix: ")
print(confusion_matrix(y_test, y_pred))
print()
print("Precision: ")
print(precision_score(y_test, y_pred)*100)
print()
print("\nClassification Report: ")
print(classification_report(y_test, y_pred))
```

**Creating the SMOTE**

```python
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report

sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

**Predicting and the fiitng the value**

```python
svm = SVC()
svm_mod = svm.fit(X_train_res, y_train_res)
y_pred = svm_mod.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: \n{accuracy * 100:.2f}")
print()
print("Confusion Matrix: ")
print(confusion_matrix(y_test, y_pred))
print()
print("Precision: ")
print(precision_score(y_test, y_pred)*100)
print()
print("\nClassification Report: ")
print(classification_report(y_test, y_pred))
```

**Importing the modules**

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from arch.bootstrap import IIDBootstrap
```

**Calling and fitting the knn model**

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

**Initialising and putting in no. of iterations.**

```python
n_iterations = 100
accuracy_scores = []
precision_scores = []
confusion_matrices = []
```

**Calling out boot-strap**

```python
boot = IIDBootstrap(X_test, y_test)
```

**Predicting the model value.**

```python
for data in boot.bootstrap(n_iterations):
    X_sample, y_sample = data[0][0], data[0][1]

    y_pred = model.predict(X_sample)

    accuracy_scores.append(accuracy_score(y_sample, y_pred))
    precision_scores.append(precision_score(y_sample, y_pred, average='weighted'))
    confusion_matrices.append(confusion_matrix(y_sample, y_pred))
```

**Calculating the mean value.**

```python
avg_accuracy = np.mean(accuracy_scores)
avg_precision = np.mean(precision_scores)
avg_conf_matrix = np.mean(confusion_matrices, axis=0)
```

**Printing the outputs.**

```python
print("Average Accuracy:", f"{avg_accuracy*100:.4f}")
print("Average Precision:", f"{avg_precision*100:.4f}")
print("Average Confusion Matrix:\n", avg_conf_matrix)
```
