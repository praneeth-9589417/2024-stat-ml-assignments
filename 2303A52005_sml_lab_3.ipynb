```python
import numpy as np
from numpy import sqrt,log,pi,exp
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/housing.csv')
data
```

```python


print('mean of latitude is :',data['latitude'].mean())
print('variance of latitude is :',data['latitude'].var(),"\n")

print('mean of total_rooms is :',data['total_rooms'].mean())
print('variance of total_rooms is :',data['total_rooms'].var(),"\n")

print('mean of housing_median_age is :',data['housing_median_age'].mean())
print('variance of housing_median_age is :',data['housing_median_age'].var(),"\n")

print('mean of total_bedrooms is :',data['total_bedrooms'].mean())
print('variance of total_bedrooms is :',data['total_bedrooms'].var())
```

QUESTION 2 :

```python
def FindMLEAndLLv(x):
  m = data['latitude'].mean()
  v = data['latitude'].var()
  mle = (1/(sqrt(2*pi*v)))*(exp(-(x-m)**2/(2*v)))
  return mle , np.log(mle)
x_values = [50,75,80]
for x in x_values:
  mle,lle = FindMLEAndLLv(x)
  print(f"the max likelyhood value of {x} is : {mle}")
  print(f"the log-likelihood value of {x} is : {lle}")
  print("\n")
```
