```python
list1 =[10,2,3,4,95,6]
list2 =[10,24,32,45,52,68]
print(list1+list2)
import numpy as np
list1 = np.array([10,2,3,4,95,6])
list2 = np.array([10,24,32,45,52,68])
print(list1+list2)
print(np.mean(list1))
print(np.mean(list2))
print(np.median(list1))
print(np.median(list2))
print(np.std(list1))
print(np.std(list2))
print(np.var(list1))
print(np.var(list2))
print(np.max(list1))
print(np.max(list2))
print(np.min(list1))
print(np.min(list2))
```

```python
m=[[1,2,3,4,9,6],[10,24,32,45,52,68]]
m1 =[[2,4,7,8,8,9],[10,24,32,45,52,68]]
m=np.array(m)
m1=np.array(m1)
print(np.add(m,m1))
print(np.subtract(m,m1))
print(np.multiply(m,m1))
print(np.divide(m,m1))
print(np.power(m,m1))
print(np.mod(m,m1))
print(np.transpose(m))
print(np.transpose(m1))
print(np.trace(m1))
print(np.trace(m))
print(np.linalg.matrix_rank(m))
print(np.linalg.matrix_rank(m1))





```
